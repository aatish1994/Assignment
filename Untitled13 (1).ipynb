```python
#replacing usa with usa
data['Country']=data['Country'].replace('UAS','USA')
```

```python
#so we can see that we have changed uas to usa
data
```

```python
#checking value counts
data['Country'].value_counts()
```

```python
#replacing the null values of column country
data['Country']=data['Country'].fillna(data['Country'].mode()[0])
```

```python
#repalcing the null values using mode
data
```

```python
#repalcing the dollar sign to blank space
data['Income.1']=data['Income.1'].str.replace('$','')
```

```python
#so we can see that dollar sign have been repplaced
data
```

```python
#repalcing the blank spaces with  null values
data['Income.1']=data['Income.1'].replace('',np.nan)
```

```python
#we can see that blank spaces have been repalced with null values
data
```

```python
#As there are outliers present so taking median instead of mean
data['Income.1']=data['Income.1'].fillna(data['Income.1'].median())
```

```python
#As there are outliers present so taking median instead of mean
data['Income.2']=data['Income.2'].fillna(data['Income.2'].median())
```

```python
#so we can see that null values have been replaced of columns Income.1 and Income.2
data
```

```python
#we can see that Income.1 data type is showing as object because dollar sign was present in it
data.info()
```

```python
#So changing the data type to float
data['Income.1']=data['Income.1'].astype(float)
```

```python
#so we can see that data type have been changed to float
data.info()
```

```python
#adding column Total column to the datafram which is a sum of Income.1 and Income.2
data['Total_Income']=data['Income.1']+data['Income.2']
```
