```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the dataset
data=pd.read_csv('raw_data.csv')
data
```

```python
#checking null values
data.isna().sum()
```

```python
#checking for row number 9
data.iloc[9]
```

```python
#replcaing the null value of row 9
data.iloc[9]=[210,'Jim Carry','1/1/1970','jim.carry@gmail.com','Male','Amazone','USA',12349.556,3652]
```

```python
#replacing the null value of column email row 3
data.loc[3, 'email'] = 'graeme.croisier@gmail.com'
```

```python
#replacing the null value of column email row 9
data.loc[6, 'email'] = 'modesta.batterson@gmail.com'
```

```python
data
```

```python
#replacing usa with usa
data['Country']=data['Country'].replace('UAS','USA')
```

```python
#so we can see that we have changed uas to usa
data
```

```python
#checking value counts
data['Country'].value_counts()
```

```python
#replacing the null values of column country
data['Country']=data['Country'].fillna(data['Country'].mode()[0])
```

```python
#repalcing the null values using mode
data
```

```python
#repalcing the dollar sign to blank space
data['Income.1']=data['Income.1'].str.replace('$','')
```

```python
#so we can see that dollar sign have been repplaced
data
```

```python
#repalcing the blank spaces with  null values
data['Income.1']=data['Income.1'].replace('',np.nan)
```

```python
#we can see that blank spaces have been repalced with null values
data
```

```python
#we can see that Income.1 data type is showing as object because dollar sign was present in it
data.info()
```

```python
#So changing the data type to float
data['Income.1']=data['Income.1'].astype(float)
```

```python
#so we can see that data type have been changed to float
data.info()
```

```python
#replacing the null value with mean
data['Income.1']=data['Income.1'].fillna(data['Income.1'].mean())
```

```python
#replacing the null value with mean
data['Income.2']=data['Income.2'].fillna(data['Income.2'].mean())
```

```python
#so we can see that null values have been replaced of columns Income.1 and Income.2
data
```

```python
#adding column Total column to the datafram which is a sum of Income.1 and Income.2
data['Total_Income']=np.ceil(data['Income.1']+data['Income.2'])
```

```python
#So we can see that Total Income column have been added to the dataframe
data
```

```python
#Bringing all the Date of Birth in same format
data['DOB']=pd.to_datetime(data['DOB']).dt.date
```

```python
#So we can see that we have converted the Date of Birth in one format
data
```

```python
#Splitting up the Employee name into first name and last name
data['First Name'] = data['Employee Full Name'].str.split().str.get(0)
data['Last Name'] = data['Employee Full Name'].str.split().str.get(1)
```

```python
#So we can see that WE have splitted the data into first name and last name
data
```

```python
#Dropping column Employee Full Name
data=data.drop(['Employee Full Name'],axis=1)
```

```python
#So we can see that Employee Full Name column have been dropped
data
```

```python
#Popping out the column first name
column_to_move=data.pop('First Name')
```

```python
#Popping out the column last name
column_to_move1=data.pop('Last Name')
```

```python
#Inserting it to position first
data.insert(1,'First Name',column_to_move)
```

```python
#Inserting it to position second
data.insert(2,'Last Name',column_to_move1)
```

```python
#So we can see that column first name and last name have been inserted in first and second postion
data
```
